```python
from torchvision.transforms.functional import resize

dot_img = img.cpu()
dot_img = resize(dot_img, size=(20, 20)).squeeze()
dot_img = (dot_img - dot_img.min()) / (dot_img.max() - dot_img.min())
dot_img.max(), dot_img.min(), dot_img.shape
```

```python
sims = (mid_outputs[-1] * dot_img).sum(dim=(1, 2)).softmax(dim=0)
sims.shape
```
